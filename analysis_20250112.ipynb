```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.cm as cm
import numpy as np

def plot_pareto_front_3d(front, actual):
    """
    Plots the Pareto front of optimal solutions for the 3D path tortuosity,
    with color representing alpha and annotations for extreme alpha and beta values.

    Args:
        front (dict): A dictionary of edge_lengths, travel_distances_to_base, and
            path_coverages for each (alpha, beta) value on the Pareto front.
        actual (tuple): The actual total_root_length, total_travel_distance, and
            total_path_coverage of the original plant.
    """
    # Extract data from the Pareto front
    alphas_betas = list(front.keys())
    data = list(front.values())
    root_lengths = [item[0] for item in data]  # Total root lengths
    travel_distances = [item[1] for item in data]  # Total travel distances
    path_coverages = [item[2] for item in data]  # Total path coverages

    # Extract the actual data
    mactual, sactual, pactual = actual

    # Extract alpha and beta values
    alphas = [alpha for alpha, beta in alphas_betas]
    betas = [beta for alpha, beta in alphas_betas]

    # Normalize alpha for colormap
    alpha_norm = plt.Normalize(min(alphas), max(alphas))
    cmap = cm.viridis

    # Identify extreme alpha and beta values
    alpha_min_idx = alphas.index(min(alphas))
    alpha_max_idx = alphas.index(max(alphas))
    beta_min_idx = betas.index(min(betas))
    beta_max_idx = betas.index(max(betas))

    # Create a 3D plot
    fig = plt.figure(figsize=(12, 9))
    ax = fig.add_subplot(111, projection="3d")

    # Plot Pareto front points
    sc = ax.scatter(
        root_lengths,
        travel_distances,
        path_coverages,
        c=alphas,
        cmap=cmap,
        alpha=0.8,
        label="Pareto Front",
    )

    # Highlight the actual point
    ax.scatter(
        mactual,
        sactual,
        pactual,
        c="red",
        s=150,
        label="Actual",
        edgecolors="black",
    )

    # Annotate extreme points
    for idx, label in zip(
        [alpha_min_idx, alpha_max_idx, beta_min_idx, beta_max_idx],
        ["Min ", "Max ", "Min ", "Max "],
    ):
        x, y, z = root_lengths[idx], travel_distances[idx], path_coverages[idx]
        alpha, beta = alphas_betas[idx]
        ax.text(
            x,
            y,
            z,
            f"{label}\n(={alpha:.2f}, ={beta:.2f})",
            fontsize=9,
            color="black",
            ha="center",
        )

    # Add a colorbar for alpha values
    cbar = plt.colorbar(sc, ax=ax, shrink=0.6, aspect=10, pad=0.1)
    cbar.set_label("Alpha Value ()", rotation=270, labelpad=15)

    # Add labels, legend, and title
    ax.set_xlabel("Total Root Length (Wiring Cost)")
    ax.set_ylabel("Total Travel Distance (Conduction Delay)")
    ax.set_zlabel("Total Path Coverage")
    ax.set_title("3D Pareto Front with Alpha (Color) and Extreme Value Annotations")

    # Show the plot
    plt.legend()
    plt.show()

```

```python
from scipy.interpolate import griddata
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def plot_pareto_surface_3d(front, actual):
    """
    Plots a 3D Pareto surface using interpolation.

    Args:
        front (dict): A dictionary of edge_lengths, travel_distances_to_base, and
                      path_coverages for each (alpha, beta) value on the Pareto front.
        actual (tuple): The actual total_root_length, total_travel_distance, and
                        total_path_coverage of the original plant.
    """
    # Extract data from the Pareto front
    alphas_betas = list(front.keys())
    data = list(front.values())
    root_lengths = np.array([item[0] for item in data])  # Total root lengths
    travel_distances = np.array([item[1] for item in data])  # Total travel distances
    path_coverages = np.array([item[2] for item in data])  # Total path coverages

    # Extract the actual data
    mactual, sactual, pactual = actual

    # Create a grid for interpolation
    grid_x, grid_y = np.meshgrid(
        np.linspace(root_lengths.min(), root_lengths.max(), 100),
        np.linspace(travel_distances.min(), travel_distances.max(), 100),
    )

    # Interpolate path_coverages onto the grid
    grid_z = griddata(
        (root_lengths, travel_distances),
        path_coverages,
        (grid_x, grid_y),
        method="cubic",
    )

    # Plot the surface
    fig = plt.figure(figsize=(12, 9))
    ax = fig.add_subplot(111, projection="3d")
    surf = ax.plot_surface(
        grid_x,
        grid_y,
        grid_z,
        cmap="viridis",
        edgecolor="k",
        alpha=0.8,
    )

    # Plot the actual point
    ax.scatter(
        mactual, sactual, pactual, color="red", s=150, label="Actual", edgecolors="black"
    )

    # Add a colorbar
    cbar = plt.colorbar(surf, ax=ax, shrink=0.5, aspect=10, pad=0.1)
    cbar.set_label("Path Coverage", rotation=270, labelpad=15)

    # Add labels and title
    ax.set_xlabel("Total Root Length (Wiring Cost)")
    ax.set_ylabel("Total Travel Distance (Conduction Delay)")
    ax.set_zlabel("Total Path Coverage")
    ax.set_title("3D Pareto Surface with Interpolated Path Coverage")

    plt.legend()
    plt.show()
```

```python
plot_pareto_surface_3d(front_3d, actual)
```
